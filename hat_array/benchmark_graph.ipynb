```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hat_df = pd.read_csv('hat_benchmark.csv')
resizable_array_df = pd.read_csv('resizable_array_benchmark.csv')
```

```python
plt.plot(hat_df['operation_amount'].head(10), hat_df['cpu_count'].head(10), label="HAT array")
plt.plot(resizable_array_df['operation_amount'].head(10), resizable_array_df['cpu_count'].head(10), label="Resizable array")
plt.xlabel('Nth operation')
plt.xlabel('CPU count')
plt.legend()
plt.grid()
plt.show()
```
