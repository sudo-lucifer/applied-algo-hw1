```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hat_df = pd.read_csv('hat_benchmark.csv')
resizable_array_df = pd.read_csv('resizable_array_benchmark.csv')

plt.plot(hat_df['operation_amount'], hat_df['cpu_count'], label="HAT array")
plt.plot(resizable_array_df['operation_amount'], resizable_array_df['cpu_count'], label="Resizable array")
plt.xlabel('Nth operation')
plt.ylabel('CPU count')
plt.legend()
plt.grid()
plt.show()
```

```python
hat_scan_df = pd.read_csv('hat_scan_benchmark.csv')
resizable_array_scan_df = pd.read_csv('resizable_array_scan_benchmark.csv')

plt.plot(hat_scan_df['read'], hat_scan_df['cpu_count'], label="HAT array")
plt.plot(resizable_array_scan_df['read'], resizable_array_scan_df['cpu_count'], label="Resizable array")
plt.xlabel('Nth read')
plt.ylabel('CPU count')
plt.legend()
plt.grid()
plt.show()
```
