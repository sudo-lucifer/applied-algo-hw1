```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
skip_list_df = pd.read_csv('skip_list_insert.csv')
linked_list_df = pd.read_csv('linked_list_insert.csv')

plt.plot(skip_list_df['operation'], skip_list_df['cpu_count'], label="Skip list")
plt.plot(linked_list_df['operation'], linked_list_df['cpu_count'], label="Linked list")
plt.xlabel('Nth operation')
plt.ylabel('CPU count')
plt.legend()
plt.grid()
plt.show()
```
